j)Construya la funci√≥n ùë¶[ùëõ]=sin (ùúã‚àó0.18ùëõ) en el intervalo 0 <= n <= 80.

```python
import matplotlib.pyplot as plt
import numpy as np

n=np.linspace(0,80,200)
y=np.sin(np.pi*0.18*n)

fig,ax=plt.subplots()
ax.set_xlabel('Tetha')
ax.set_ylabel('Tiempo')
ax.set_title("y(n)=sin (pi‚àó0.18n)")
ax.plot(n,y, c='c')
ax.grid(True)
plt.show()
```

k) Construya otra senoidal ùë¶2[ùëõ]=cos(2ùúã‚àó0.03ùëõ)

```python
n2=np.linspace(0,80,200)
y2=np.cos(2*np.pi*0.03*n2)

fig,ax=plt.subplots()
ax.set_xlabel('Tetha')
ax.set_ylabel('Tiempo')
ax.set_title("y2(n)=cos (2pi‚àó0.03n)")
ax.plot(n2,y2, c='b')
ax.grid(True)
plt.show()
```

l)Genere una tercera se√±al que sea la suma de estas dos se√±ales, es decir ùë†[ùëõ]=y[n]+ùë¶2[ùëõ] y una cuarta, que sea el producto de estas, ùë°[ùëõ]=y[n].ùë¶2[ùëõ].

```python
s=y+y2
fig,ax=plt.subplots()
ax.set_xlabel('Tetha')
ax.set_ylabel('Tiempo')
ax.set_title("s(n)=y[n]+y2[n]")
ax.plot(s, c='y')
ax.grid(True)
plt.show()
```

```python
t=y*y2
fig,ax=plt.subplots()
ax.set_xlabel('Tetha')
ax.set_ylabel('Tiempo')
ax.set_title("t(n)=y(n).y2(n)")
ax.plot(t, c='r')
ax.grid(True)
plt.show()
```

# ***¬°RETO!***
Escribir una funci√≥n que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota m√≠nima, la m√°xima, media y la desviaci√≥n t√≠pica.

```python
import numpy as np
import pandas as pd

def resumen_notas(diccionario_notas):
    # La funci√≥n recibe las notas en forma de diccionario
    notas = pd.Series(diccionario_notas)
    #Convierte el diccionario en una serie de pandas
    # Utiliza las funciones para calcular la nota m√≠nima, m√°xima, media y la desviaci√≥n est√°ndar

    resumen = pd.Series({
        ' La m√≠nima nota es': notas.min(),
        'La m√°xima nota es': notas.max(),
        'La media de los estudiantes es': notas.mean(),
        'La desviaci√≥n de las notas es': notas.std()
    })

    return resumen

# Ejemplo de uso
notas_alumnos = {'Alumno1': 3.5,'Alumno2': 4.0,'Alumno3': 4.5,'Alumno4': 4.7,'Alumno5': 2.5}

resumenestudiantes = resumen_notas(notas_alumnos)
print(resumenestudiantes)
```
