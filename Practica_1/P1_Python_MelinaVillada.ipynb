j)Construya la función 𝑦[𝑛]=sin (𝜋∗0.18𝑛) en el intervalo 0 <= n <= 80.

```python
import matplotlib.pyplot as plt
import numpy as np

n=np.linspace(0,80,200)
y=np.sin(np.pi*0.18*n)

fig,ax=plt.subplots()
ax.set_xlabel('Tetha')
ax.set_ylabel('Tiempo')
ax.set_title("y(n)=sin (pi∗0.18n)")
ax.plot(n,y, c='c')
ax.grid(True)
plt.show()
```

k) Construya otra senoidal 𝑦2[𝑛]=cos(2𝜋∗0.03𝑛)

```python
n2=np.linspace(0,80,200)
y2=np.cos(2*np.pi*0.03*n2)

fig,ax=plt.subplots()
ax.set_xlabel('Tetha')
ax.set_ylabel('Tiempo')
ax.set_title("y2(n)=cos (2pi∗0.03n)")
ax.plot(n2,y2, c='b')
ax.grid(True)
plt.show()
```

l)Genere una tercera señal que sea la suma de estas dos señales, es decir 𝑠[𝑛]=y[n]+𝑦2[𝑛] y una cuarta, que sea el producto de estas, 𝑡[𝑛]=y[n].𝑦2[𝑛].

```python
s=y+y2
fig,ax=plt.subplots()
ax.set_xlabel('Tetha')
ax.set_ylabel('Tiempo')
ax.set_title("s(n)=y[n]+y2[n]")
ax.plot(s, c='y')
ax.grid(True)
plt.show()
```

```python
t=y*y2
fig,ax=plt.subplots()
ax.set_xlabel('Tetha')
ax.set_ylabel('Tiempo')
ax.set_title("t(n)=y(n).y2(n)")
ax.plot(t, c='r')
ax.grid(True)
plt.show()
```

# ***¡RETO!***
Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota mínima, la máxima, media y la desviación típica.

```python
import numpy as np
import pandas as pd

def resumen_notas(diccionario_notas):
    # La función recibe las notas en forma de diccionario
    notas = pd.Series(diccionario_notas)
    #Convierte el diccionario en una serie de pandas
    # Utiliza las funciones para calcular la nota mínima, máxima, media y la desviación estándar

    resumen = pd.Series({
        ' La mínima nota es': notas.min(),
        'La máxima nota es': notas.max(),
        'La media de los estudiantes es': notas.mean(),
        'La desviación de las notas es': notas.std()
    })

    return resumen

# Ejemplo de uso
notas_alumnos = {'Alumno1': 3.5,'Alumno2': 4.0,'Alumno3': 4.5,'Alumno4': 4.7,'Alumno5': 2.5}

resumenestudiantes = resumen_notas(notas_alumnos)
print(resumenestudiantes)
```
