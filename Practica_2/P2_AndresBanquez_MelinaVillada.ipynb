### Impulso unitario

El impulso unitario se define como:

En Python, implementamos el impulso unitario mediante la siguiente funci√≥n:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = [-2, -1, 0, 1, 2, 3]
x = [5, 6, 7, 5, 4, 3]
```

```python
def impseq(n0, n1, n2):
# Genera x(n) = delta(n-n0); n1 <= n <= n2
# ----------------------------------------------
 n = np.arange(n1,n2+1) # Se crea el vector de muestras
 x = (n-n0) == 0
 return [x,n]
```

```python
x,n = impseq(5, 0, 10)
```

```python
plt.stem(n,x)
plt.title("Funci√≥n Impulso Unitario")
plt.ylabel("Amplitud")
```

##Escalon Unitario

1. Implemente una funci√≥n en Python que permita generar una funci√≥n escal√≥n unitario, definida en un intervalo ùëõ1‚â§ùëõ0‚â§ùëõ2

```python
def escalon(n0, n1, n2):
  n = np.arange(n1, n2+1)
  x = (n>=n0) == 1
  return [x, n]
```

```python
x, n = escalon(5,0,10)
plt.stem(n,x)
plt.title("Funci√≥n escal√≥n")
plt.ylabel("Amplitud")
```

###Rampa

2. Implemente una funci√≥n en Python que permita generar una funci√≥n rampa, definida en un intervalo ùëõ1‚â§ùëõ0‚â§ùëõ2 con una pendiente m.

```python
def rampa(n0, n1, n2, m = 1):
  u,n = escalon(n0, n1, n2)
  x=u*(n-n0)*m
  return [x, n]
```

```python
x, n= rampa(1, 0, 5, m = 3)
plt.stem(n,x)
plt.title("Funci√≥n rampa")
plt.ylabel("Amplitud")
```

###Exponenciales complejas
Las exponenciales complejas son de la forma:

image.png

En la que œÉ produce una atenuaci√≥n si es negativo o una amplificaci√≥n si es positivo. En Python se hace
uso del comando np.exp

```python
def exponencial(n1: int, n2: int, sigma: float = 1):
  n = np.arange(n1,n2,1)
  x = np.exp(sigma * n)
  return x, n
```

```python
x, n = exponencial(0,10, sigma = 1)
plt.stem(n, x)
plt.title("Funci√≥n exponencial")
plt.ylabel("Amplitud")
```

### Sinusoides

Estas secuencias son de la forma:

image.png

```python
def sinusoides(n1, n2, omega0 = 0, w = 1):
  n = np.arange(n1,n2,1)
  x = np.sin(w*n + omega0)
  return x, n
```

```python
x, n = sinusoides(0,10, omega0 = np.pi/4)
plt.stem(n, x)
plt.title("Funci√≥n sinusoide")
plt.ylabel("Amplitud")
```

```python
[x1,n1]=rampa(0,0,5);
[x2,n2]=rampa(5,6,11);
[x3,n3]=rampa(10,12,17);
#Se realiza la concatenaci√≥n
X1=np.concatenate((x1,x2,x3))
N1 = np.concatenate((n1,n2,n3))
#Se grafica la funci√≥n:
plt.stem(N1,X1)
plt.title("Secuencia punto 4")
plt.ylabel("Amplitud")
```

a. Derive la sencuencia

```python
# Para derivar secuencias se usa el comando np.diff()
derivx1 = np.diff(x1)
derivx2 = np.diff(x2)
derivx3 = np.diff(x3)
derivadatotal = np.concatenate((derivx1, derivx2, derivx3))
nderiv=np.concatenate((n1[:-1], n2[:-1], n3[:-1])) #Se debe ajustar la longitud 
plt.stem(nderiv, derivadatotal)
```
